```python
selected_columns = [
    'loan_amnt', 'term', 'int_rate', 'installment', 'grade', 'sub_grade', 
    'emp_title', 'emp_length', 'home_ownership', 'annual_inc', 
    'verification_status', 'issue_d', 'loan_status', 'purpose', 'title', 
    'zip_code', 'addr_state', 'dti', 'earliest_cr_line', 'open_acc', 
    'pub_rec', 'revol_bal', 'revol_util', 'total_acc', 'initial_list_status', 
    'application_type', 'mort_acc', 'pub_rec_bankruptcies'
]
```

```python
loan_data_subset = loan_data[selected_columns]
```

```python
##Data Pre-processing
for column in loan_data_subset.columns:
    if loan_data_subset[column].isna().sum() != 0:
        missing = loan_data_subset[column].isna().sum()
        portion = loan_data_subset[column].isna().mean() * 100
        print(f"'{column}': number of missing values '{missing}' ==> '{portion:.3f}%'")
```

```python
loan_data_subset['emp_title'].nunique()
```

```python
#Drop the emp_title column
loan_data_subset.drop('emp_title', axis =1, inplace=True)
```

```python
#Drop the title column
loan_data_subset.drop('title', axis =1, inplace=True)
```

```python
corrs = loan_data_subset.corr(numeric_only = True)['mort_acc'].drop('mort_acc').sort_values()
sns.barplot(x=corrs,y=corrs.index)
plt.title('Correlation of Features with mort_acc')
plt.show()
```

```python
loan_data_subset['mort_acc'].unique()
```

```python
#filling missing values in mort_acc( filling the mean value group by total_acc, 
#Logically, a borrower with more total credit accounts (higher total_acc) is more likely to have at least one mortgage account.)
loan_data_subset['mort_acc'] = loan_data_subset['mort_acc'].fillna(
    loan_data_subset.groupby('total_acc')['mort_acc'].transform('median'))

loan_data_subset['mort_acc'].isnull().sum()
```

```python
for column in loan_data_subset.columns:
    if loan_data_subset[column].isna().sum() != 0:
        missing = loan_data_subset[column].isna().sum()
        portion = loan_data_subset[column].isna().mean() * 100
        print(f"'{column}': number of missing values '{missing}' ==> '{portion:.3f}%'")
```

```python
#Drop the emp_length column
loan_data_subset.drop('emp_length', axis =1, inplace=True)
```

```python
#dropping the na values in the data as the percentage the na is very low in all the columns
loan_data_subset.dropna(inplace = True)
```

```python
#Drop the grade column as it looks like a superset of sub_grade column
loan_data_subset.drop('grade', axis =1, inplace=True)
```

```python
#Drop issue_d as this is irrelevant for the predictive model
loan_data_subset.drop('issue_d', axis =1, inplace=True)
```

```python
#extracting year from earliest_cr_line
loan_data_subset['earliest_cr_line'] = pd.to_datetime(loan_data_subset['earliest_cr_line'],format = '%b-%Y',errors = 'coerce')
loan_data_subset['earliest_cr_line'] = loan_data_subset['earliest_cr_line'].dt.year
```

```python
loan_data_subset['earliest_cr_line'].unique()
```

```python
loan_data_subset.head(100).T
```

```python
## changing term values to int ##
term_dict = {' 36 months':36,' 60 months':60}
loan_data_subset['term'] = loan_data_subset['term'].map(term_dict)
```

```python
loan_data_subset.to_csv('loan_data_latest.csv', index=False)
```

```python
dummies = ['sub_grade', 'verification_status', 'purpose', 'initial_list_status', 
           'application_type', 'home_ownership']
loan_data_subset = pd.get_dummies(loan_data_subset, columns=dummies, drop_first=True)
```

```python
#drop zip_code from dataset
loan_data_subset.drop('zip_code', axis =1, inplace=True)
```

```python
loan_data_subset['addr_state'].unique()
```

```python
state_to_region = {
    # Northeast
    'CT':'Northeast', 'MA':'Northeast', 'ME':'Northeast', 'NH':'Northeast',
    'NJ':'Northeast', 'NY':'Northeast', 'PA':'Northeast', 'RI':'Northeast', 'VT':'Northeast',

    # Midwest
    'IA':'Midwest', 'IL':'Midwest', 'IN':'Midwest', 'KS':'Midwest', 'MI':'Midwest',
    'MN':'Midwest', 'MO':'Midwest', 'ND':'Midwest', 'NE':'Midwest', 'OH':'Midwest',
    'SD':'Midwest', 'WI':'Midwest',

    # South
    'AL':'South', 'AR':'South', 'DC':'South', 'DE':'South', 'FL':'South',
    'GA':'South', 'KY':'South', 'LA':'South', 'MD':'South', 'MS':'South',
    'NC':'South', 'OK':'South', 'SC':'South', 'TN':'South', 'TX':'South',
    'VA':'South', 'WV':'South',

    # West
    'AK':'West', 'AZ':'West', 'CA':'West', 'CO':'West', 'HI':'West',
    'ID':'West', 'MT':'West', 'NM':'West', 'NV':'West', 'OR':'West',
    'UT':'West', 'WA':'West', 'WY':'West'
}

loan_data_subset['region'] = loan_data_subset['addr_state'].map(state_to_region)
```

```python
loan_data_subset.drop('addr_state',axis=1,inplace=True)
```

```python
loan_data_subset = pd.get_dummies(loan_data_subset, columns=['region'], drop_first=True)
```
